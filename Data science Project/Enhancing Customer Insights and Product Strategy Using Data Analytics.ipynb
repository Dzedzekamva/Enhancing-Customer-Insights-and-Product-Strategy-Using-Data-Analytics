## Handle missing values by filling them with appropriate statistics (mean, median) or removing them, depending on the context.

```python
# Checking missing values
Data.isnull().sum()
```

```python
# Impute missing values with mode for each column
for column in Data1.columns:
    mode_value = Data1[column].mode()[0]  # [0] is used to get the first mode in case of multiple modes
    Data1[column].fillna(mode_value, inplace=True)
```

```python
# Display the DataFrame after mode imputation
print("\nDataFrame after Mode Imputation:")
Data1
```

## Correct data types for each column (e.g., converting strings to numeric values where appropriate).

```python
# Check if 'PRICE' column is numeric
def is_numeric(column):
    try:
        pd.to_numeric(column)
        return True
    except ValueError:
        return False
```

```python
# Check if 'PRICE' column is numeric
is_numeric_price = is_numeric(Data1['PRICE'])
print(f"Is 'PRICE' column numeric? {is_numeric_price}")
```

```python
Data1['PRICE'] = pd.to_numeric(Data1['PRICE'].replace('[^0-9\.-]', '', regex=True))
Data1
```

```python
# Defining a function to format phone numbers
def format_phone_number(phone):
    # Remove non-numeric characters
    numeric_phone = ''.join(c for c in phone if c.isdigit())
    
    # Formating phone number to a form (XXX) XXX-XXXX)
    formatted_phone = '({}) {}-{}'.format(numeric_phone[:3], numeric_phone[3:6], numeric_phone[6:])
    
    return formatted_phone

# Applying the format_phone_number function to the 'PHONE' column
Data1['PHONE'] = Data1['PHONE'].apply(format_phone_number)
Data1
```

```python
# Removing leading spaces for all columns
Data1.columns = Data1.columns.str.strip()
Data1
```

## Standardize the format of text data, such as customer names and product names, to ensure consistency.

```python
# Converting 'Name' to title case
Data1['NAME'] = Data1['NAME'].str.title()
Data1
```

```python
# converting product name to upper case
Data1['PRODUCTS'] = Data1['PRODUCTS'].str.upper()
Data1
```
